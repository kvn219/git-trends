```python
import pandas as pd
import requests
```

```python
url = "https://api.github.com/search/repositories?order=desc&per_page=100&q=data+language%3Ago&sort=stars"
response = requests.get(url, )
body = response.json()
results = body['items']
for i, r in enumerate(results):
    print(f"{i}\n{r['description']}\n{r['html_url']}\n")
```
